```python
import pandas as py
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
```

```python
dt = py.read_csv('data.csv')
```

```python
dt
```

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
```

```python
imputer.fit(dt.iloc[:,1:3].values)
```

```python
dt.iloc[:,1:3] = imputer.transform(dt.iloc[:,1:3].values)
```

```python
dt
```

now same for country

```python
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imputer.fit(dt.iloc[:,:1].values)
dt.iloc[:,:1] = imputer.transform(dt.iloc[:,:1].values)
```
