```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

Scatterplot, in numerical data when we have to find the relationship, scatterplot helps us a lot

```python
sns.scatterplot(x = "age", y="fare",data=df)
```

```python
sns.scatterplot(x = "age", y="fare",data=df, hue = "survived")
```

```python
df2 = sns.load_dataset("tips")
```

```python
df2.head()
```

```python
sns.scatterplot(x="total_bill", y="tip",data=df2, hue = "smoker")
```

```python
sns.scatterplot(x="total_bill", y="tip",data=df2, hue = "sex")
```

```python
sns.scatterplot(x="total_bill", y="tip",data=df2, hue = "sex", style = "smoker")
```

Jointplot, it is needed when we need distribution as well as scatter plot

```python
sns.jointplot(x = "total_bill", y = "tip", data = df2)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df2, hue = "sex")
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = df2, kind = "reg" )
```

Now we will use the lineplot, it is basically used when we want to find the relation between 2 variables and also trend over time

```python
df3 = sns.load_dataset("dowjones")
```

```python
df3.head()
```

```python
sns.lineplot(x = "Date", y = "Price", data = df3)
```

```python
df3["Date"] = pd.to_datetime(df3["Date"])
df3["Month"] = df3["Date"].dt.month

df3["Year"] = df3["Date"].dt.year
```

```python
df3.head()
```

```python
sns.lineplot(x = "Date", y = "Price", data = df3[df3["Year"]==1925])
```
