```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

```python
sns.countplot(x="survived", data = df)
```

```python
sns.countplot(x="survived", data = df, hue = "class")
```

now what will happen if we have more number of values of more categories of x axis so it is becomes more tough to analyze data using bargraph, so we use heatmap

so first we will be making pivot table 



```python
x = pd.crosstab(df["survived"], df["class"])
```

```python
x
```

```python
sns.heatmap(x)
```

now i want value written in the boxes

```python
sns.heatmap(x, annot = True)
```

```python
sns.heatmap(x, annot = True, fmt = ".1f")
```

```python
sns.heatmap(x, annot = True, fmt = ".1f", cmap = "coolwarm")
```

now, if we have more than 2 category, we will use barplot, but it is important that you have a numerical value for barplot


```python
sns.barplot(x = "class", y= "age", hue="sex", data = df)
```

If you want the distribution that how are the values distributed, then you use strip plot

```python
sns.stripplot(x="class", y="age", hue = "sex", data = df)
```
