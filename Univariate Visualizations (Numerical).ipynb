```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df
```

Histogram

```python
sns.histplot(data = df, x = "price")
```

```python
sns.histplot(data = df[df["price"]<1000], x = "price")
```

```python
sns.histplot(data = df[df["price"]<500], x = "price")
```

```python
sns.histplot(data = df, x = "availability_365")
```

now we have large number of values, so we use distplot coz it gives probability distribution of dataset

```python
sns.distplot(df["price"])
```

```python
sns.distplot(df["availability_365"])
```

```python
sns.distplot(df["reviews_per_month"])
```

boxplot gives us outliers and where the min, max values lie in the data set, if we have a small dataset then boxplots are very much essential

```python
sns.boxplot(y=df["price"])
```

```python
sns.boxplot(y=df["availability_365"])
```
