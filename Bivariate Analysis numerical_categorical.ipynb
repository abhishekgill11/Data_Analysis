Bivariate analysis helps us to visualise and find out the statistical link between any 2 variables. so, in this we are going to take 1 numerical and 1 categorical data to find the relation between them using bivariate analysis

```python
import pandas as pd
import numpy as np
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
```

```python
df.head()
```

now if we have numerical and categorical data, we can plot a bargraph

```python
sns.barplot(x = "class", y="age", data = df)
```

in documentation it is written that we can use estimator function with this, in estimator we can give anything so we will give sum here instead of mean as written in the documentation

```python
sns.barplot(x = "class", y="age", data = df, estimator = "sum")
```

```python
sns.barplot(x = "class", y="age", data = df, estimator = np.sum)
```

here the sum does not have any importance, but if i use fare instead of age then it will give the total sum of fare in each class as well

```python
sns.barplot(x = "class", y="fare", data = df, estimator = np.sum)
```

now we also have a hue value in documentation, so what heue does is it categorises the value in further categories

```python
sns.barplot(x = "class", y="age", data = df, hue = "sex")
```

now it divided age in according to the gender

now we can use hue on fare so that it will give us that what is the maximum fare we are getting from each gender

```python
sns.barplot(x = "class", y="fare", data = df, hue = "sex", estimator = np.sum)
```

```python
sns.boxplot(x = "class", y="age", data = df)
```

now we know that this is a box plot, so this blue,orange,green indicates iqr

```python
sns.boxplot(x = "class", y="age", data = df, hue = "survived")
```

now, sometimes when we use a boxplot, it may require a distribution as well so for that we will use violin plot

```python
sns.violinplot(x = "class", y="age", data = df)
```

now we can check for the particular distribution as well using distplot

```python
sns.distplot(df[df["class"] == "Second"]["age"])
```
