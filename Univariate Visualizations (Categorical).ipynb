```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head(2)
```

Categorical

Bar Graph

```python
sns.countplot(data=df, x = "neighbourhood_group")
```

```python
sns.countplot(data=df, x = "room_type")
```

how many hotels have how many different room types in each city

```python
sns.countplot(data=df, x="neighbourhood_group", hue = "room_type")
```

Pie chart
we use pie chart when we want that particular percentage value in the whole category

```python
df["room_type"].value_counts().plot(kind = "pie")
```

```python
df["room_type"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["neighbourhood_group"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["room_type"].value_counts().plot(kind = "bar")
```
